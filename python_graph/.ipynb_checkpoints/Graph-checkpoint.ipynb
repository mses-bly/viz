```python
from functools import reduce
class Graph(object):
    def __init__(self, file = None):
        self.N = 0
        self.G = {}
        if file is not None:
            self.N = int(file.readline()[:-1])
            self.G = {i: set() for i in range(self.N)}
            for line in file:
                edge = line.rstrip('\n').split()
                print(edge)
                self.add_edge_bi(int(edge[0]), int(edge[1]))
            file.close()
            
            
        
    def add_edge(self, s, t):
        _set = self.G[s]
        _set.add(t)
        self.G[s] = _set
    
    def add_edge_bi(self, s, t):
        self.add_edge(s,t)
        self.add_edge(t,s)
        
    def adj(self, s):
        return self.G[s]
    
    def _print(self):
        print(self.G)
        
    def json(self):
        nodes = []
        
        for i in range(self.N):
            node = """{"id":"%s"}""" % i
            nodes.append(node)
        
        edges = []
        for source in range(self.N):
            for target in self.adj(source):
                edge = """{"source":"%s", "target":"%s"}""" % (source, target)
                edges.append(edge)
                
        nodes_str = reduce((lambda x, y: """%s, %s""" % (x, y)), nodes)
        edges_str = reduce((lambda x, y: """%s, %s""" % (x, y)), edges)
        
        return """{"nodes": [%s], "links": [%s]}""" % (nodes_str, edges_str)
            
    
#         {
#   "nodes": [
#     {
#       "name": "node 1",
#       "artist": "artist name",
#       "id": "unique_id_1",
#       "playcount": 123
#     },
#     {
#       "name": "node 2",
#       # ...
#     }
#   ],
#   "links": [
#     {
#       "source": "unique_id_1",
#       "target": "unique_id_2"
#     },
#     {
#       # ...
#     }
#   ]
# }

g = Graph(open('graph.txt', 'r'))
g._print()


out = open('graph_json.txt', 'w')
out.write("%s\n" % g.json())

# g.add_edge_bi(3, 4)
# print(g.json())
```
